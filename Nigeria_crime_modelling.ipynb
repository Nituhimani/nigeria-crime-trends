
# Training and finding the best model

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Loading data
data = pd.read_csv('/content/Nigeria_1997-2024_Sep20_cleaned.csv')

# Define features and target variable
X = data.drop('fatalities', axis=1)
y = data['fatalities']

# Identify columns with object (string) and datetime dtypes
object_cols = X.select_dtypes(include=['object', 'datetime']).columns

# Create a LabelEncoder object
label_encoder = LabelEncoder()

# Applying label encoding to each object and datetime column
for col in object_cols:
    # Convert datetime to string before label encoding
    if pd.api.types.is_datetime64_any_dtype(X[col]):
        X[col] = X[col].astype(str)
    X[col] = label_encoder.fit_transform(X[col])

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define the models and their pipelines
pipelines = {
    "Logistic Regression": Pipeline([('scaler', StandardScaler()), ('lr', LogisticRegression())]),
    "Decision Tree": Pipeline([('scaler', StandardScaler()), ('dt', DecisionTreeClassifier())]),
    "Random Forest": Pipeline([('scaler', StandardScaler()), ('rf', RandomForestClassifier())]),
    "Gradient Boosting": Pipeline([('scaler', StandardScaler()), ('gb', GradientBoostingClassifier())]),
    "SVM": Pipeline([('scaler', StandardScaler()), ('svc', SVC())])
}

# Train and evaluate each model
results = {}
for name, pipeline in pipelines.items():
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    results[name] = accuracy
    print(f"{name} Accuracy: {accuracy:.4f}")  # Print accuracy for each model

# Find the best model
best_model = max(results, key=results.get)
best_accuracy = results[best_model]

print(f"\nBest Model: {best_model}")
print(f"Accuracy: {best_accuracy:.4f}")
```
